```python
from openai import OpenAI
import os
from dotenv import load_dotenv
import fitz
from ebooklib import epub
import ebooklib
from bs4 import BeautifulSoup


load_dotenv()
```

Setup chatgpt

```python
# Set up API key

client = OpenAI(api_key=os.environ.get("OPENAI_API_KEY"))

chat_completion = client.chat.completions.create(
    messages=[
        {
            "role": "user",
            "content": "Say this is a test",
        }
    ],
    model="gpt-3.5-turbo",
)

print(chat_completion.choices[0].message.content)
```

Import pdf and convert to text

```python
def pdf_to_text(pdf_path):
    # Open the PDF file
    document = fitz.open(pdf_path)
    
    text = ""
    # Iterate through each page
    for page_num in range(len(document)):
        page = document.load_page(page_num)
        text += page.get_text()

    return text

# Example usage
pdf_path = '../books/pdfs/The_Picture_of_Dorian_Gray.pdf'  # Replace with your PDF file path
text = pdf_to_text(pdf_path)
print(text)
```

Convert from epub to txt

```python
def epub_to_text(epub_file):
    book = epub.read_epub(epub_file)
    text_content = []

    for item in book.get_items_of_type(ebooklib.ITEM_DOCUMENT):
        # Use BeautifulSoup to parse HTML content and extract text
        soup = BeautifulSoup(item.get_content(), 'html.parser')
        text = soup.get_text(separator='\n', strip=True)
        text_content.append(text)

    return_text = ""
    for text in text_content:
        return_text += text
        return_text += '\n\n'  # Add double newline between documents (if multiple)
    
    return return_text

# Example usage:
epub_file = '../books/epubs/Dear_Leader.epub'

text = epub_to_text(epub_file)
print(text)
```
